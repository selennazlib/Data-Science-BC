K Means Clustering is an unsupervised algorithm that will attempt to group similar clusters together.

There is no easy answer for choosing a best K values. One way is the elbow method. (Plot K against the SSE(sum squared error) -> the error decreases as k gets larger) 

The idea of the elbow method is to choose the K at which the SSE decreases abruptly.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
data = make_blobs(n_samples=200, n_features=2, centers=4,
                  cluster_std=1.8, random_state=101)
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1], cmap='rainbow')
```

```python
from sklearn.cluster import KMeans

kmm = KMeans(n_clusters=4)
```

```python
kmm.fit(data[0])
```

```python
kmm.cluster_centers_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 6))

ax1.set_title('KMeans')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmm.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
# n_cluster -> 2
kmm = KMeans(n_clusters=2)
kmm.fit(data[0])

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 6))

ax1.set_title('KMeans')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmm.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
# n_cluster -> 10
kmm = KMeans(n_clusters=10)
kmm.fit(data[0])

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 6))

ax1.set_title('KMeans')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmm.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```
